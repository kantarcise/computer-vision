## One Intermediate Step - Analog Cameras ‚ú®

Cameras are tools that freeze a single moment.

And before we had digital cameras, we had analog cameras which were based on the same principles of camera obscura.

![Film Camera](../img/digital_cameras/film_camera.jpg)

**‚ÄúAnalogue‚Äù** comes from a Greek word meaning ‚Äúin proportion‚Äù or ‚Äúsimilar.‚Äù 

In the case of cameras, "analog" means that the entire camera captures an image of the film that is a continuous, seamless representation of the world. Similar to the way our eyes view the world. 

It is like a drawing or painting made up of many small pieces connected together.

Analog cameras, in their simplest form, consist of 5 parts.

* **Lens**: It is the round cam part located on the front of the camera. It helps focus light from the Earth onto a special film at the core of the camera.
* **Aperture**: Aperture is an opening that primarily limits light propagated through the camera.
* **Film**: This is like a special paper that captures the picture. It's inside the camera and when you take a photo, light from the earth hits the film and a photo is created!
* **Shutter**: It is like a curtain that opens and closes to let in light. When you press the button to take a photo, the shutter opens and then closes again.
* **Film Roll**: . Where the photo was saved. After a photo, it moves a bit to prepare for the next photo.

## How exactly a digital image is made? ü§î

- Light from the thing you are photographing zooms into the camera lens. 
- This incoming "picture" hits the image sensor chip, which breaks it up into millions of pixels. 
- The sensor measures the color and brightness of each pixel and stores it as a number. 
- Your digital photograph is effectively an enormously long string of numbers describing the exact details of each pixel it contains.

## Image Sensors

These are **semiconductor devices** that records light electronically. 

![Image Sensor](../img/digital_cameras/canoneosc70Inside.jpg)

[Source](https://www.canon.com.tr/pro/infobank/image-sensors-explained/)

There are 2 main types of image sensors used today.

In the past, the image sensor employed by most digital cameras was a charge coupled device (CCD). 

The surface of a CCD Sensors is packed with photodiodes, each of which senses light and accumulates electrical charge in accordance with the strength of light that strikes it.

They were used in NASA's interplanetary spacecraft. But they had problems when exposed to radiation and power dissipation was high. To use the CCD there were a lot of other parts needed (like ADC) that were bulky and power-hungry. [Source](https://ericfossum.com/Publications/Papers/2023%20IEEE%2075th%20Anniversary%20of%20the%20Transistor%20-%202023%20-%20Nathan%20-%20The%20Invention%20and%20Development%20of%20CMOS%20Image%20Sensors.pdf)

To address these problems, another approach was suggested, the CMOS!

Most of cameras today use complementary metal oxide semiconductor (CMOS) technology instead. [Source](https://electronics.howstuffworks.com/cameras-photography/digital/digital-camera1.htm)

[Recommended read](https://ericfossum.com/Publications/Papers/2023%20IEEE%2075th%20Anniversary%20of%20the%20Transistor%20-%202023%20-%20Nathan%20-%20The%20Invention%20and%20Development%20of%20CMOS%20Image%20Sensors.pdf) about CMOS image sensors.

## Power of the Customer! ‚úÖ

With the promise of **lower power consumption** and **higher integration for smaller components**, CMOS designers focused efforts on imagers for mobile phones, the highest volume image sensor application in the world.

## Capturing Color

A limitation of individual photosites is that they are unable to perceive color. 

Instead, they only measure the **overall brightness** of the light that hits them. To capture a full-color image, most camera sensors employ a filtering system that breaks down light into its three primary colors. 

By recording the intensity of each of these colors, the camera can then combine them to produce a complete and vibrant color spectrum.

### One Solution - Beam Splitter 

A beam splitter can be used to receive color! 

To illustrate this process, imagine light entering the camera as water flowing through a pipe. 

The beam splitter acts as a divider, channeling an identical amount of "water" into three separate pipes, each leading to a distinct sensor. 

Although each sensor receives an identical view of the image, the filters ensure that each sensor only responds to one of the three primary colors.

The key benefit of this approach is that the camera captures all three colors at every pixel location, resulting in a more accurate and detailed color representation. 

However, a drawback of this method is that it often leads to cameras that are larger in size and more expensive. [Source](https://electronics.howstuffworks.com/cameras-photography/digital/digital-camera4.htm)

![Beam Splitter](../img/digital_cameras/beam_splitter.jpg)

### Another Solution - Filters!

A more cost-effective and practical approach to capturing primary colors is to affix a color filter array directly over each individual photosite. 

By dividing the sensor into a mosaic of red, blue, and green pixels, it's possible to gather sufficient information from the surrounding area to make highly accurate estimates of the true color at each location.

This process, known as **interpolation**, involves analyzing the data from neighboring pixels and making an informed inference about the color at a given point ([Source](https://electronics.howstuffworks.com/cameras-photography/digital/digital-camera5.htm)).

![Filters](../img/digital_cameras/demosaicing.png)

[Source](https://www.bitcraze.io/2020/07/ai-deck-sees-in-color/)

## Why is there a lot of green in Bayer ? ü§î

The Bayer filter pattern is 50% green, 25% red and 25% blue. 

The reason there is more green is that both silicon and the human eye are **most responsive around the green** wavelengths.

![Bayer Filter](../img/digital_cameras/bayerFilter.png)

[Source](https://en.wikipedia.org/wiki/Bayer_filter)

The raw output of Bayer-filter cameras is referred to as a **Bayer pattern image**. 

Since each pixel is filtered to record only one of three colors, the data from each pixel cannot fully specify each of the red, green, and blue values on its own. 

To obtain a full-color image, various **demosaicing** algorithms can be used to interpolate a set of complete red, green, and blue values for each pixel. 

These algorithms make use of the surrounding pixels of the corresponding colors to estimate the values for a particular pixel.

Multiple cameras, provides different features.

The **ultrawide camera** gives you the ability to shoot at $0.5x$ with a wider field of view.

The **telephoto camera** give you the ability to make optical zoom.

## What kind of filter does the iPhone use ? ü§î

### Quad Bayer Filter on iPhone 15

The Quad Bayer structure means four adjacent pixels are clustered with the same-color filters.

What really changes is not the filter but the sensor behind it ‚Äì these new sensors put four pixels behind each color square instead of just one. You can read more about it [here](https://www.gsmarena.com/quad_bayer_sensors_explained-news-37459.php)

In this way, high sensitivity and high resolution can be achieved in one sensor. 

It can, for example, prevent resolution loss in a low-illuminance environment and produce low-noise nightscape photos.
